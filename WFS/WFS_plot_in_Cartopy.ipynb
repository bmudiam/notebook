# Read WFS data from USGS ScienceBase into Shapely/Cartopy

```python
import geojson
import requests
from owslib.wfs import WebFeatureService
from shapely.geometry import Polygon, mapping, asShape
import cartopy.crs as ccrs
import cartopy.feature
from cartopy.io.img_tiles import MapQuestOpenAerial, MapQuestOSM, OSM
```

```python
#endpoint='https://www.sciencebase.gov/catalogMaps/mapping/ows/5342c54be4b0aa151574a8dc'
endpoint='https://www.sciencebase.gov/catalogMaps/mapping/ows/5342c5fce4b0aa151574a8ed'
wfs = WebFeatureService(endpoint)
print wfs.version
```

```python
shp = wfs.contents.keys()
print shp
```

```python
# Read shapefile data from USGS ScienceBase WFS 1.0 service in JSON format
# (shapefile was previosly uploaded to ScienceBase, creating the WFS service)
url = ('%s?service=wfs&version=1.0.0&request=GetFeature&'
'typeNames=%s&outputFormat=application/json') % (endpoint, shp[1])
```

```python
geo = mapping(geojson.loads(requests.get(url).text))
```
